```python
# Importar las librerías necesarias
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Crear el dataset (valores de abdomen y antena para cada insecto)
data = {
    'abdomen': [2.7, 0.9, 1.1, 2.9, 0.5, 8.3, 8.1, 8.4, 5.4, 6.1],
    'antena': [5.5, 4.7, 3.1, 1.9, 1.5, 4.7, 9.1, 6.6, 8.5, 6.6],
    'clase': ['saltamontes', 'saltamontes', 'saltamontes', 'saltamontes',
              'saltamontes', 'chicharra', 'chicharra', 'chicharra', 'chicharra', 'chicharra']
}

# Convertir a un DataFrame de pandas
df = pd.DataFrame(data)
print(df)
```

```python
# Definir X (características) e Y (etiquetas)
X = df[['abdomen', 'antena']]  # Características
Y = df['clase']  # Etiquetas

# Dividir los datos en conjunto de entrenamiento y conjunto de prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Crear el modelo de SVM
model = SVC()

# Entrenar el modelo
model.fit(X_train, Y_train)
```

```python
# Realizar predicciones en el conjunto de prueba
Y_pred = model.predict(X_test)

# Evaluar el modelo con la precisión
accuracy = accuracy_score(Y_test, Y_pred)
print(f'Precisión del modelo: {accuracy * 100:.2f}%')

# Mostrar las predicciones
print(f'Predicciones: {Y_pred}')
```

```python
# Hacer una predicción con un nuevo dato
new_data = [[5.1, 7.0]]  # Nuevo tamaño de abdomen y antena
prediction = model.predict(new_data)
print(f'La predicción para el nuevo insecto es: {prediction[0]}')
```
