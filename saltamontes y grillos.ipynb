```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Datos de entrenamiento
X_train = np.array([[2.5, 5.0], [3.0, 6.0], [4.0, 7.5], [3.5, 5.5]])
y_train = np.array(['Saltamontes', 'Saltamontes', 'Chicharra', 'Chicharra'])

# Entrenar el modelo SVM
model = svm.SVC(kernel='linear')
model.fit(X_train, y_train)

# Crear gráfico
plt.scatter(X_train[:, 0], X_train[:, 1], c=['blue' if label == 'Saltamontes' else 'green' for label in y_train])
plt.xlabel('Tamaño del abdomen (cm)')
plt.ylabel('Tamaño de la antena (cm)')
plt.title('Clasificación de Saltamontes y Chicharras con SVM')
plt.show()

# Predicción con un nuevo ejemplo
X_test = np.array([[3.8, 6.0]])  # Nuevo insecto
prediccion = model.predict(X_test)
print("La predicción para el insecto es:", prediccion[0])
```
